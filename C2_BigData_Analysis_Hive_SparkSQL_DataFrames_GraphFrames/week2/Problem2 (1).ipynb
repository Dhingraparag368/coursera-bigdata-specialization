Compare most popular tags in year 2016 with tags in 2009. Tag popularity is the number of questions (post_type_id=1) with this tag. Output top 10 tags in format:



php 5 3 1234 5678

```python
%%writefile query_2.hql

ADD JAR /opt/cloudera/parcels/CDH/lib/hive/lib/hive-contrib.jar;
ADD JAR /opt/cloudera/parcels/CDH/lib/hive/lib/hive-serde.jar;
USE stackoverflow_;
with 
B_2016 as
(select tag,cnt as cnt_2016,RANK() OVER (ORDER BY cnt desc) as rnk_2016 from A where year='2016'),
B_2009 as
(select tag,cnt as cnt_2009,RANK() OVER (ORDER BY cnt desc) as rnk_2009 from A where year='2009'),
A as(
select year,tag,count(*) as cnt
    from posts LATERAL VIEW explode(tags) exploded as tag 
    where year in ('2016','2009')  group by year,tag
)
select B_2016.tag,rnk_2016,rnk_2009,cnt_2016,cnt_2009 from 
B_2016 left outer join B_2009
on B_2016.tag=B_2009.tag
where rnk_2016<11
order by rnk_2016 
;
```
